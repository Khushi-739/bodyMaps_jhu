```python
import os

base = "/kaggle/working/SupremProject/AbdomenAtlasDemo"

for folder in os.listdir(base):
    path = os.path.join(base, folder)

    if os.path.isdir(path):

        # remove mac hidden files
        for f in os.listdir(path):
            if f.startswith("._"):
                os.remove(os.path.join(path, f))

        # rename ct.nii → ct.nii.gz
        if os.path.exists(os.path.join(path, "ct.nii")):
            os.rename(
                os.path.join(path, "ct.nii"),
                os.path.join(path, "ct.nii.gz")
            )

print("Dataset fixed")
```
